```python
# Question 9: 
# A multiple-choice test consists of 10 questions, each with four possible answers. If astudent randomly guesses on each question, what is the probability of gettingat least
# 8 questions correct? Data: Number of questions (n) = 10, Number of possible
# answers per question (k) = 4

from scipy.stats import binom

n = 10
p = 0.25

prob = binom.pmf(8, n, p) + binom.pmf(9, n, p) + binom.pmf(10, n, p)

print("Probability:", prob)
print("""
Interpretation:
- The probability of getting at least 8 correct answers by random guessing is extremely low
- This means it is very unlikely to score high marks by guessing
""")
```

```python
# Question 10:
# The lifetimes of a certain brand of light bulbs are normally distributed witha meanof 1000 hours and a standard deviation of 100 hours.
# What is the probabilitythat arandomly selected light bulb lasts between 900 and 1100 hours? 
# Data: Meanlifetime(μ) = 1000 hours, Standard deviation (σ) = 100 hours, Lifetime range (lower limit x1, upper limit x2)
from scipy.stats import norm

mean = 1000
std = 100

prob = norm.cdf(1100, mean, std) - norm.cdf(900, mean, std)

print("Probability:", prob)
```

```python
#  Question 11:
#  A study is conducted to estimate the mean height of a population. Arandomsampleof 100 individuals is selected, and their heights are measured.
#  Calculate a 95%confidence interval for the population mean height, given that the sample meanheight is 170 cm and the sample standard deviation is 8 cm. 
#  Data: Sample size(n) =100, Sample mean (x̄) = 170 cm, Sample standard deviation (s) = 8 cm, Confidencelevel = 95%
import math

mean = 170
std = 8
n = 100
z = 1.96

margin = z * (std / math.sqrt(n))

lower = mean - margin
upper = mean + margin

print("Confidence Interval:", lower, "to", upper)
```

```python
# # Question 12:
# A manufacturing company claims that the average weight of its product is 500grams. To test this claim, a random sample of 25 products is selected, and their
# weights are measured. The sample mean weight is found to be 510 grams withasample standard deviation of 20 grams. Perform a hypothesis test to determineif
# there is evidence to support the company's claim. Data: Sample size (n) =25, Sample mean (x̄) = 510 grams, Sample standard deviation (s) = 20 grams, Population mean (μ) = 500 grams
x_bar = 510
mu = 500
s = 20
n = 25

z = (x_bar - mu) / (s / math.sqrt(n))

print("Z-score:", z)

# Decision
if abs(z) > 1.96:
    print("Reject H0")
else:
    print("Fail to Reject H0")
    
print("""
Here our Z-value is greater than the significance level of alpha (2.5 > 1.96),
so we reject the null hypothesis and conclude that the sample mean is
significantly different from the population mean.
""")    
    
```

```python
# Question 14:
# A survey is conducted to estimate the proportion of people in a city who support aparticular policy. A random sample of 500 individuals is surveyed, and 320of themexpress support for the policy.
# Calculate a 90% confidence interval for the population proportion, given the sample proportion. Data: Sample size (n) =500, Number of successes (x) = 320, Confidence level = 90%

n = 500
x = 320

p = x / n
z = 1.645

se = math.sqrt((p * (1 - p)) / n)
margin = z * se

lower = p - margin
upper = p + margin

print("Confidence Interval:", lower, "to", upper)
```

```python
# Question 15:
# A company sells smartphones, and the number of defects per batch follows a Poisson distribution with a mean of 2 defects. 
# What is the probability of havingexactly 3 defects in a randomly selected batch? Data: Mean number of defects (λ) =2, Number of defects (x) = 3

lam = 2
x = 3

prob = (math.exp(-lam) * lam**x) / math.factorial(x)

print("Probability:", prob)
```
